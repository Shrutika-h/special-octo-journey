```
import pandas as pd
import matplotlib.pyplot as plt

path = "/content/drive/My Drive/Rainfall prediction/Sub_Division_IMD_2017.csv"
data = pd.read_csv(path)
```

```

data = data.drop(['ANNUAL','JF','MAM','JJAS','OND'],axis=1)
data_s=data.melt(id_vars=["SUBDIVISION","YEAR"])
#data_s.info

data_s


```

```
data_s = data_s.assign(MONTH=pd.to_datetime(data_s.variable, format='%b').dt.month).drop(['variable'],axis=1)
data_s = data_s.assign(Date=pd.to_datetime(data_s[['YEAR', 'MONTH']].assign(day=1)))
final_data = data_s.drop(['YEAR','MONTH'],axis=1)
main_data = pd.pivot_table(final_data, values='value', columns=['SUBDIVISION'],index=["Date"])
```

```
#saving the transformed data to csv file
main_data.to_csv("/content/drive/My Drive/Rainfall prediction/final_data.csv")
```

```
main_data = main_data.drop(['Andaman & Nicobar Islands','Arunachal Pradesh','Lakshadweep'],axis=1)
#main_data.isnull().sum()
main_data = main_data.reset_index()
main_data = main_data.fillna(main_data.mean(),axis=0)
```
